### Performing Exploratory Data Analysis on Titanic data set to gain insights about the scenario 

```python
#Importing Usable Libraries

import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
td_df = pd.read_csv(r"C:\Users\reals\Downloads\titanic.csv")
td_df.head(10)
```

## Cleaning the data 

```python
td_df.shape
```

Shape of the data is 891 Rows and 12 Columns

```python
td_df.info()
```

```python
td_df.isnull().sum()
```

There are Null values in the Age column , Cabin column and Embarked column

```python
# Replacing AGe columns Null Values with its median
Age_med = td_df["Age"].median()
td_df['Age'].fillna(Age_med,inplace=True)
```

```python
# Dropping the Cabin Column Because it has Alot of Null VAlue's and its not neccessary at all
td_df.drop(columns={'Cabin'},inplace=True)
```

```python
# Embarked column had only two null values so we fill them with their mode
Embarked_mod = td_df['Embarked'].mode()[0]  # Get the mode value itself, not the Series
td_df['Embarked'].fillna(Embarked_mod, inplace=True)
```

```python
# Lets Check now if we have any missiing values 
td_df.isnull().sum()
```

We removed/replaced every null value from the data set 

Now we can perform EDA

# Exploratory Data Analysis

```python
td_df.head()
```

```python
# Shows how many peoples survived and how many lost their life 
survival_counts = td_df['Survived'].value_counts()
print(survival_counts)
```

```python
td_df['Survived'].value_counts(normalize=True).plot(kind='bar', color='skyblue')
plt.title('Survival Rate')
plt.show()
```

```python
sns.histplot(data=td_df, x='Age', bins=30, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count of People')
plt.show()
```

```python
td_df['Survived'] = td_df['Survived'].astype(str)
data_label = sns.countplot(data=td_df, x='Sex', hue='Survived', palette=['red', 'blue'])
data_label.bar_label(data_label.containers[0])
data_label.bar_label(data_label.containers[1])
plt.title('Survival Rate by Sex')
plt.ylabel('Count')
plt.xlabel('Sex')
plt.show()
```

```python
data_label = sns.countplot(data=td_df, x='Pclass', hue='Survived', palette=['red', 'blue'])
data_label.bar_label(data_label.containers[0])
data_label.bar_label(data_label.containers[1])
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.show()
```

```python
sns.histplot(data=td_df, x='Age', hue='Survived', bins=30, palette=['red', 'green'], alpha=0.6)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.show()
```

```python
sns.pairplot(td_df, hue='Survived', vars=['Age', 'Fare']) 
```

```python
# Shows relation  between numeric variables of data 

numeric_td = td_df.drop(columns={'Name','Sex','Ticket','Embarked'})

corr_matrix = numeric_td.corr()

cax = plt.imshow(corr_matrix, interpolation='nearest', cmap='coolwarm')
plt.colorbar(cax)
plt.title('Correlation Matrix')
plt.xticks(np.arange(len(corr_matrix.columns)), corr_matrix.columns, rotation=45, ha='right')
plt.yticks(np.arange(len(corr_matrix.columns)), corr_matrix.columns)

for i in range(len(corr_matrix.columns)):
    for j in range(len(corr_matrix.columns)):
        plt.text(j, i, f'{corr_matrix.iloc[i, j]:.2f}', ha='center', va='center', color='black')

plt.tight_layout()
plt.show()

```
